```python
import numpy as np
```

```python
from keras.models import Sequential
from keras.layers import LSTM
from keras.layers import Dense
```

```python
#Sequence length
length = 5
```

```python
seq = np.array([i/float(length) for i in range(length)])#0/5,1/5,2/5,3/5,4/5
print(seq)
```

```python
X = seq.reshape(len(seq),1,1)#The input for LSTMs must be three dimensional
Y = seq.reshape(len(seq),1)
print(X.shape)#sequence with 5 samples, 1 time step, and 1 feature.
print(Y.shape)
```

```python
n_neurons = length
n_batch = length
n_epoch = 1000
```

# model 

```python
model = Sequential()
model.add(LSTM(n_neurons,input_shape = (1,1)))#5 neurons,1 input 
model.add(Dense(1))# 1 output
```

```python
model.compile(optimizer = 'adam',loss = 'mean_squared_error')
```

```python
model.summary()
```

```python
model.fit(X,Y,epochs = n_epoch,batch_size = n_batch)
```

```python
#Evaluate
result =  model.predict(X,batch_size = n_batch,verbose = 2)
```
